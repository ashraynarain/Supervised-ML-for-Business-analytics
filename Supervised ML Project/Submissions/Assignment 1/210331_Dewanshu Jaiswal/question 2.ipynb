```python
import matplotlib
x = [1,2,3,4,5,6]
y = [4,7,10,13,16,19]
matplotlib.pyplot.plot(x,y)
matplotlib.pyplot.title("Graph of straight line : y=3x+1")
matplotlib.pyplot.xlabel("X")
matplotlib.pyplot.ylabel("Y")
matplotlib.pyplot.show()

import matplotlib
x=['A','E','C','D']
y=[1,25,9,16]
matplotlib.pyplot.bar(x,y,color='red',width=0.8)
matplotlib.pyplot.title("Bar Graph")
matplotlib.pyplot.xlabel("letter")
matplotlib.pyplot.ylabel("square of position of letter")
matplotlib.pyplot.show()

import matplotlib
favourite_fruit=['apple','mango','banana','watermelon']
number_of_persons=[5,10,8,2]
matplotlib.pyplot.pie(number_of_persons,labels=favourite_fruit,shadow=True,radius=1)
matplotlib.pyplot.title("Pie Chart")
matplotlib.pyplot.show()
```
